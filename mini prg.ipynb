```python
import pandas as pd
df = pd.read_csv('house_price.csv')

# Helper function to convert "Posted X ago" into a number of days
def convert_verification_date(date_str):
    match = re.search(r'Posted (\d+) (day|month|year)s? ago', date_str)
    if match:
        value = int(match.group(1))
        unit = match.group(2)
        if unit == 'day':
            return value
        elif unit == 'month':
            return value * 30
        elif unit == 'year':
            return value * 365
    elif "Posted a month ago" in date_str:
        return 30
    elif "Posted a year ago" in date_str:
        return 365
    elif "Posted a day ago" in date_str:
        return 1
    return None

# Apply the conversion
df['verificationDate'] = df['verificationDate'].apply(lambda x: convert_verification_date(str(x)))

# Now process the five tasks

# 1. Average of verificationDate (in days)
average_verification_days = df['verificationDate'].mean()

# 2. Total sum of the price column
total_price = df['price'].sum()

# 3. Count of each value in the city column
city_counts = df['city'].value_counts()

# 4. Sum of verificationDate for each unique city
verification_sum_by_city = df.groupby('city')['verificationDate'].sum()

# 5. Highest verificationDate (i.e., the oldest post)
max_verification_days = df['verificationDate'].max()

average_verification_days, total_price, city_counts, verification_sum_by_city, max_verification_days
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import numpy as np

# Load the dataset
df = pd.read_csv("house_price.csv")  # Update the path if needed

# Extract 'Age' from 'verificationDate'
def extract_age(text):
    match = re.search(r"(\d+)\s+year", text)
    if match:
        return int(match.group(1))
    if "a month ago" in text.lower():
        return 0
    return np.nan

df["Age"] = df["verificationDate"].astype(str).apply(extract_age)

# Drop rows with missing Age values
df_cleaned = df.dropna(subset=["Age"])

# Set seaborn style
sns.set(style="whitegrid")

# 1. Bar Plot: City vs. Average Age
plt.figure(figsize=(8, 5))
avg_age_by_city = df_cleaned.groupby('city')['Age'].mean()
avg_age_by_city.plot(kind='bar', color='skyblue')
plt.title('Average Age by City')
plt.ylabel('Average Age (Years)')
plt.xlabel('City')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Line Plot: Age over Index
plt.figure(figsize=(8, 5))
plt.plot(df_cleaned['Age'].reset_index(drop=True), marker='o', linestyle='-', color='green')
plt.title('Age over Index')
plt.ylabel('Age (Years)')
plt.xlabel('Index')
plt.tight_layout()
plt.show()

# 3. Histogram: Age
plt.figure(figsize=(8, 5))
sns.histplot(df_cleaned['Age'], bins=10, kde=True, color='orange')
plt.title('Distribution of Age')
plt.xlabel('Age (Years)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 4. Pie Chart: City Distribution
plt.figure(figsize=(6, 6))
city_counts = df_cleaned['city'].value_counts()
plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('City Distribution')
plt.tight_layout()
plt.show()

# 5. Scatter Plot: Age vs. Price
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_cleaned, x='Age', y='price', hue='city')
plt.title('Scatter Plot of Age vs. Price')
plt.xlabel('Age (Years)')
plt.ylabel('Price')
plt.legend(title='City')
plt.tight_layout()
plt.show()
```

```python
from datetime import datetime, date
# Initialize Faker
fake = Faker()

# Function to calculate age from date of birth
def calculate_age(dob):
    today = date.today()  # Use current date dynamically
    return today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))

# Generate fake dates of birth and calculate ages
def generate_fake_data(num_records):
    data = []
    for _ in range(num_records):
        dob = fake.date_of_birth(minimum_age=18, maximum_age=80)
        age = calculate_age(dob)
        data.append({"Date of Birth": dob, "Age": age})
    return pd.DataFrame(data)

# Generate 100 records and display them
fake_data = generate_fake_data(100)
print(fake_data)
```
